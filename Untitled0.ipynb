```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
from imblearn.over_sampling import SMOTE
import xgboost as xgb
import seaborn as sns
import matplotlib.pyplot as plt

train_df = pd.read_csv('/content/train.csv')
test_df = pd.read_csv('/content/test.csv')

print(train_df.info())
print(train_df.head())
print(train_df.describe())
```

```python
# Handle missing values in cigarettes_per_day
train_df['cigarettes_per_day'] = train_df['cigarettes_per_day'].fillna(0)
test_df['cigarettes_per_day'] = test_df['cigarettes_per_day'].fillna(0)
```

```python
print("Train missing values:\n", train_df.isnull().sum())
print("Test missing values:\n", test_df.isnull().sum())
```

```python
le = LabelEncoder()
categorical_cols = ['sex', 'family_cancer_history', 'has_other_cancer', 'asthma_diagnosis', 'liver_condition', 'blood_pressure_status']
for col in categorical_cols:
    train_df[col] = le.fit_transform(train_df[col])
    test_df[col] = le.transform(test_df[col])
```

```python
train_df = pd.get_dummies(train_df, columns=['smoking_status', 'treatment_type', 'residence_state'], drop_first=True)
test_df = pd.get_dummies(test_df, columns=['smoking_status', 'treatment_type', 'residence_state'], drop_first=True)
```

```python
test_df = test_df.reindex(columns=train_df.columns.drop(['survival_status', 'record_id']).tolist() + ['record_id'], fill_value=0)
```

```python
train_df['treatment_start_date'] = pd.to_datetime(train_df['treatment_start_date'])
train_df['treatment_end_date'] = pd.to_datetime(train_df['treatment_end_date'])
test_df['treatment_start_date'] = pd.to_datetime(test_df['treatment_start_date'])
test_df['treatment_end_date'] = pd.to_datetime(test_df['treatment_end_date'])
train_df['treatment_duration_days'] = (train_df['treatment_end_date'] - train_df['treatment_start_date']).dt.days
test_df['treatment_duration_days'] = (test_df['treatment_end_date'] - test_df['treatment_start_date']).dt.days
```

```python
reference_date = pd.to_datetime('2025-07-14')
train_df['diagnosis_date'] = pd.to_datetime(train_df['diagnosis_date'])
test_df['diagnosis_date'] = pd.to_datetime(test_df['diagnosis_date'])
train_df['days_since_diagnosis'] = (reference_date - train_df['diagnosis_date']).dt.days
test_df['days_since_diagnosis'] = (reference_date - test_df['diagnosis_date']).dt.days
```

```python
train_df['bmi'] = train_df['weight_kg'] / (train_df['height_cm'] / 100) ** 2
test_df['bmi'] = test_df['weight_kg'] / (test_df['height_cm'] / 100) ** 2
```

```python
train_df = train_df.drop(['diagnosis_date', 'treatment_start_date', 'treatment_end_date'], axis=1)
test_df = test_df.drop(['diagnosis_date', 'treatment_start_date', 'treatment_end_date'], axis=1)
```

```python
scaler = StandardScaler()
numerical_cols = ['patient_age', 'height_cm', 'weight_kg', 'cigarettes_per_day', 'cholesterol_mg_dl', 'cancer_stage', 'treatment_duration_days', 'days_since_diagnosis', 'bmi']
train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])
test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])
```

```python
X = train_df.drop(['record_id', 'survival_status'], axis=1)
y  = train_df['survival_status']
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_val)
print("Accuracy:", accuracy_score(y_val, y_pred))
print("F1 Score:", f1_score(y_val, y_pred))
print(classification_report(y_val, y_pred))
```

```python
X_test = test_df.drop('record_id', axis=1)
```

```python
test_predictions = model.predict(X_test)
```

```python
submission = pd.DataFrame({
    'record_id': test_df['record_id'],
    'survival_status': test_predictions
})
submission.to_csv('submission.csv', index=False)
```
